```python
import pandas as pd
import os
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
folder_path = r'.\ml-1m' 
unames = ['user_id', 'gender', 'age', 'occupation', 'zip'] 
users = pd.read_table(os.path.join(folder_path,'users.dat'), sep='::', header=None, names=unames)
rnames = ['user_id', 'movie_id', 'rating',	'timestamp'] 
ratings = pd.read_table(os.path.join(folder_path,'ratings.dat'), sep = '::', header=None, names=rnames) 
mnames = ['movie_id', 'title', 'genres'] 
movies = pd.read_table(os.path.join(folder_path,'movies.dat'),sep='::',header=None,	names=mnames,engine='python', encoding='ISO-8859-1')
```

```python
print(users[:5])
print(ratings[:5])
print(movies[:5])
```

## Content-based filtering

```python
from scipy.sparse import csr_matrix
movie_type_value=[]
type_total=[]
for _,row in movies.iterrows():
    type_split=row['genres'].split('|')
    for i in type_split:
        if i not in type_total:
            type_total.append(i)
        movie_type_value.append((row['movie_id'],type_total.index(i),1))
print(type_total)
print(movie_type_value)
num_movie = max([x[0] for x in movie_type_value]) +1
num_type = max([x[1] for x in movie_type_value]) +1
zip_movie, zip_type, zip_values = zip(*movie_type_value)
sparse_matrix = csr_matrix((zip_values, (zip_movie, zip_type)), shape=(num_movie, num_type))
```

```python
# This is a sparse matrix indicating that the second type appeared once in the first movie ---> (0, 2) 1
print(sparse_matrix.shape)
print(sparse_matrix)
```

```python
#Cosine similarity [-1,1] between (i,j)
cosine_sim = cosine_similarity(sparse_matrix)
print(cosine_sim.shape)
print(cosine_sim)
```

```python
def content_based_recommendation(user_id,number):

    #The movies reviewed by user_id.
    user_movies = ratings[ratings['user_id'] == user_id]
    movie_indices = user_movies['movie_id'].values
    # Calculate the sum of similarities between each movie and the movies reviewed. 
    scores = cosine_sim[movie_indices].sum(axis=0)
    recommended_movie_indices = sorted(range(len(scores)), key=lambda i: scores[i], reverse=True)
    #print(user_movies)
    #print((movie_indices))
    #print(len(scores))3883
    #print(recommended_movie_indices)
    recommended_movie_names=[]
    for index,i in enumerate(recommended_movie_indices):
        if(index==number):
            break
        recommended_movie_names.append(movies.loc[movies['movie_id'] == i, 'title'].values[0])
    return recommended_movie_names
user_id = 1
number = 5
print(content_based_recommendation(user_id,number))
```

## Item-based collaborative filtering

```python
movie_user_value=[]
for _,row in ratings.iterrows():
    movie_user_value.append((row['movie_id'],row['user_id'],row['rating']))
num_movies = max([x[0] for x in movie_user_value]) +1
num_user = max([x[1] for x in movie_user_value]) +1
print(num_movies,num_user)
print(movie_user_value[0])
zip_movie, zip_users, zip_values = zip(*movie_user_value)
sparse_matrix = csr_matrix((zip_values, (zip_movie, zip_users)), shape=(num_movies, num_user))    
print(sparse_matrix.shape)
print(sparse_matrix)
```

```python
cosine_sim = cosine_similarity(sparse_matrix)
print(cosine_sim.shape)
print(cosine_sim)
```

```python
def Item_based_collaborative_filtering_ranking_list(user_id):
    movie_temp = ratings[ratings['user_id'] == user_id]
    movie_indices = movie_temp['movie_id'].values
    weight_user_dict={}
    for _ in movie_indices:
        weight_user_dict[_]=ratings.loc[(ratings['user_id'] == user_id) & (ratings['movie_id'] == _), 'rating'].values[0]
    print(weight_user_dict)
    user_weight_list=[]
    user_weight_list.append(-1)#The 0th dimension represents no ratings; there is no 'Movie 0'.
    for i in range(1,len(cosine_sim)):
        #print(i)
        _=0#sum
        __=0#devide
        for j in movie_indices:
            _+=round(cosine_sim[i][j]*weight_user_dict[j],5)#*ratings.loc[(ratings['user_id'] == user_id) & (ratings['movie_id'] == j), 'rating'].values[0]
            __+=round(cosine_sim[i][j],5)
        if i in weight_user_dict:
            #print(i)
            user_weight_list.append(weight_user_dict[i])
        else:
            if __==0:
                user_weight_list.append(0)
            else:
                user_weight_list.append(_/__)
    return  user_weight_list
user_id=2

Item_based_collaborative_filtering_ranking_list_for_user_id = Item_based_collaborative_filtering_ranking_list(user_id)
print(Item_based_collaborative_filtering_ranking_list_for_user_id)
print(len(Item_based_collaborative_filtering_ranking_list_for_user_id))  
sorted_list_with_indices = sorted(enumerate(Item_based_collaborative_filtering_ranking_list_for_user_id), key=lambda x: x[1], reverse=True)
sorted_values = [value for index, value in sorted_list_with_indices]
original_indices_sorted = [index for index, value in sorted_list_with_indices]
print("The scores after sorting.", sorted_values)
print("The indices corresponding to the scores.", original_indices_sorted)
name_sorted=[]
for i in original_indices_sorted:
    name_sorted.append(str(movies[(movies['movie_id'] == i)]["title"].values))
print(name_sorted)
```

## User-based collaborative filtering

```python
user_movie_value=[]
for _,row in ratings.iterrows():
    user_movie_value.append((row['user_id'],row['movie_id'],row['rating']))
num_user = max([x[0] for x in user_movie_value]) +1
num_movie = max([x[1] for x in user_movie_value]) +1
print(user_movie_value[0])
```

```python
zip_user, zip_movie, zip_values = zip(*user_movie_value)
sparse_matrix = csr_matrix((zip_values, (zip_user, zip_movie)), shape=(num_user, num_movie))    
print(sparse_matrix.shape)
print(sparse_matrix)
```

```python
cosine_sim = cosine_similarity(sparse_matrix)
print(cosine_sim.shape)
print(cosine_sim[1])
print(cosine_sim)
```

```python
def User_based_collaborative_filtering_list(user_id,number,each_recommend):
    sorted_user = sorted(enumerate(cosine_sim[user_id]), key=lambda x: x[1], reverse=True)
    sorted_user_values = [value for index, value in sorted_user]
    original_user_sorted = [index for index, value in sorted_user]
    res=set()
    for i in range(1,1+number):
        movie_temp = ratings[ratings['user_id'] == original_user_sorted[i]]
        movies_indices = movie_temp['movie_id'].values
        ratings_indices = movie_temp['rating'].values
        sorted_list_with_indices = sorted(zip(movies_indices,ratings_indices), key=lambda x: x[1], reverse=True)
        #print(sorted_list_with_indices)
        sorted_movie = [movie for movie, rating in sorted_list_with_indices]
        print(f'The {i}th most similar recommendation: {sorted_movie}')
        res.update(sorted_movie[:5])
    print(f'All recommendations:{res}')
    alread_have=set()
    movie_temp = ratings[ratings['user_id'] == user_id]
    movie_indices = movie_temp['movie_id'].values
    print(f'The {user_id}th user has already watched:{movie_indices}')
    for i in movie_indices:
        alread_have.add(i)
    #print(sorted_user_values)
    #print(original_user_sorted)
    return res-alread_have
user_id=2
number=2
each_recommend=5
recommend_set=User_based_collaborative_filtering_list(user_id,number,each_recommend)
print(f'The indices of movies to recommend to {user_id}: {recommend_set}')
name_sorted=[]
for i in recommend_set:
    name_sorted.append(str(movies[(movies['movie_id'] == i)]["title"].values))
print(name_sorted)
```
